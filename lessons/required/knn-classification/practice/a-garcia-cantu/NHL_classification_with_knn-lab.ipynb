### 10. Fit the KNN at values of K from 1 to the number of rows in the training data.
- Store the test accuracy in a list.
- Plot the test accuracy vs. the number of neighbors.

```python
# A:

test_accuracy = []
for k in range(1, X_train.shape[0]+1):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    test_accuracy.append(knn.score(X_test, y_test))
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(list(range(1, X_train.shape[0]+1)), test_accuracy, lw=3.)
plt.show()
```

### 11. Fit KNN across different values of K and plot the mean cross-validated accuracy with 5 folds.

```python
# A:

from sklearn.model_selection import cross_val_score

max_neighbors = np.floor(X.shape[0] - X.shape[0]/5.)

print(max_neighbors)
```

```python
test_accuracy = []
for k in range(1, int(max_neighbors)):
    knn = KNeighborsClassifier(n_neighbors=k)
    test_accuracy.append(np.mean(cross_val_score(knn, X, y, cv=5)))
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(list(range(1, int(max_neighbors))), test_accuracy, lw=3.)
plt.show()
```
